```python
#cifras significativas
import numpy as np #Se importa la libreria

x = np.linspace(0,2,1000)
#print(x)

f = x**2-2
x_aprox = x[np.argmin(np.abs(f))] #np.agrmin ---> Esta función es útil para encontrar la posición del elemento más pequeño de un array.
print(x_aprox)
#03/02/26
```
