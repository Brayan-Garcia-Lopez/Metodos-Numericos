```python
#VISUALIZACION DEL ERROR
import numpy as np
import matplotlib.pyplot as plt
# Se genera un arreglo de 100 puntos en el intervalo [-1, 5]
x = np.linspace(-1,5,100)
#se define la funcion
y = (x - 2)**2

#Grafica la funcion
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Funcion de costo")
plt.show()
```
