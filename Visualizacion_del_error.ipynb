```python
#VISUALIZACION DEL ERROR
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,5,100)
y = (x - 2)**2


plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Funcion de costo")
plt.show()
```
