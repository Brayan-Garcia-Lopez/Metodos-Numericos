```python
#Resolver sistema con numpy
import numpy as np #Importar la libreria numpy para calculos numericos

A = np.array([[10,1,-1], #Matriz de coeficientes del sistema de ecuaciones
              [2,10,1], #segunda fila de coeficientes
              [2,3,10]]) #Tercera fila de coeficientes

b = np.array([11,12,13]) #vector de terminos independientes

x = np.linalg.solve(A,b) #Funcion de Numpy que resuelve el sistema de ecuaciones
print("Solucion del sistema: ", x) #Imprime el vector solucion del sistema
```

```python
#metodo de JACOBBI
x = np.zeros(3) #Vector inicial de aproximacion (x0,y0,z0)

tol = 1e-6 #tolerancia para el criterio de para
max_iter = 100 #Numero maximo de iteraciones permitidas

for k in range(max_iter): # Bucle principal de iteraciones
    x_new = np.zeros(3) # Vector donde se guarda la nueva aprocimacion

    x_new[0] = (11 - x[1] - x[2])/10 #Despeje x apartir de la primera ecuacion
    x_new[1] = (12 - 2*x[0] - x[2])/10 #Despeje y apartir de la segunda ecuacion
    x_new[2] = (13 - 2*x[0] - 3*x[1])/10 #Despeje z apartir de la tercera ecuacion

    if np.linalg.norm(x_new - x) < tol: #Verifica si el cambio de iteraciones es menor que
        break #Terminar el proceso si se cumple para la siguiente iteracion

    q = x_new #Actualiza la aproximacion para la siguiente iteracion

print("Solucion aproximada: ", x_new) #Imprime la solucion aproximada
print("Iteraciones:", k+1) #Imprime el numero de iteraciones

```

```python
#metodo de GAUSS-SEIDEL
x = np.zeros(3)

for k in range(max_iter): # Bucle principal de iteraciones
    x_old = x.copy()# Guarda la saloucion anterior para calcular el error

    x_new[0] = (11 - x[1] - x[2])/10 #Despeje x apartir de la primera ecuacion
    x_new[1] = (12 - 2*x[0] - x[2])/10 #Despeje y apartir de la segunda ecuacion
    x_new[2] = (13 - 2*x[0] - 3*x[1])/10 #Despeje z apartir de la tercera ecuacion

    if np.linalg.norm(x - x_old) < tol: #Verifica si el cambio de iteraciones es menor que
        break #Terminar el proceso si se cumple para la siguiente iteracion

    q = x_new #Actualiza la aproximacion para la siguiente iteracion

print("Solucion Gauss-Seidel: ", x_new) #Imprime la solucion aproximada
print("Iteraciones:", k+1) #Imprime el numero de iteraciones
```
