```python
#OPTIMIZACION NUMERICA
from scipy.optimize import minimize #-- Ofrece más funciones de utilidad para optimización, estadísticas y procesamiento de señales.

#La funcion
f = lambda x: (x - 2)**2 #--Lambda: sirve para crear funciones anónimas pequeñas y rápidas en una sola línea de código

#Se ejecuta el algoritmo de optimizacion
resultado = minimize(f,x0=0)

print(resultado.x)

#En IA: Ajustes automaticos parametros
```
