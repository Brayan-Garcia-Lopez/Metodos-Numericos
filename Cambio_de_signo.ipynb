```python
#Verificar cambio de signo
import numpy as pd
def f(x):
  return x**3 - x - 2

a = 1 # Limite inferior del intervalo
b = 2 # Limite superior del intervalo

fa = f(a) #Evaluaciion de la funcion en el extremo izquierdo
fb = f(b) #Evaluaciion de la funcion en el extremo derecho

print("f(a): ", fa) #Mostrar valor en a
print("f(b): ", fb) #Mostrar valor en b

if fa*fb < 0: # Verifica condicion del teorema del valor intermedio
  print("Existe al menos una raiz en el intervalo") # Confirmacion
else:
  print("No se garantiza raiz en el intervalo") # No hay cambio de signo
```

```python
#Implementar biseccion manual

def f(x): # Definicion de la funcion a analizar
  return x**3 - x - 2 # Funcion no lineal

a = 1
b = 2
tolerancia = 1e-6 # Error maximo permitido
max_iteracion = 23 # Numero maximo de iteraciones permitida

for i in range(max_iteracion): # ciclo de iteraciones
  c = (a + b) / 2 # Calculo de punto medio
  fc = f(c) #Evaluacion de la funcion en el punto medio

  if abs(fc) < tolerancia: #Criterio de paro por valor cercano a cero
    break # Termina el ciclo

  if f(a) * fc < 0: # Verifica is la raiz esta en el sub indice
    b = c # Actualiza el extremo derecho
  else:
    a = c #actualiza el extremo izquierdo

print("Raiz aproximada: ", c) #Muestra la raiz encontrada
print("Iteraciones realizadas: ", i+1) # Numero de iteraciones realizadas
```

```python
#visualizacion

a = 1 #Limite inicial inferior
b = 2 #Limite inicial superior
n = 22 #numero de iteraciones

error_teorico = (b - a) / (2**n) #Formula del error maximo en la biseccion
print("Erro maximo despues de ", n , "Iteraciones: ", error_teorico) #Mostrar error
```
