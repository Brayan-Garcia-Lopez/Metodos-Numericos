```python
#ERROR DE REDONDEO
#suma de numeros decimales
a = 0.1 + 0.2
#Valor que se espera de la suma
b = 0.3

#Imprimir la suma
print(a)
#Compara la suma con el valor esperado
print(a == b)
#Resta los valores a y b y abs por si sale negativo
print(abs(a - b))
```
