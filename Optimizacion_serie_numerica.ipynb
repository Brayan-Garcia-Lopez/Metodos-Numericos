```python
#Minimizar f(x) = (x-3)^2
#optimizacion serie numerica
#Valor inicial
x = 0
#Tasa de aprendizaje
alpha = 0.1

#Iteraciones
for _ in range(20):

    #Gradiente: es un operador vectorial que representa la máxima tasa de cambio de una función escalar en varias variables.
    grad = 2 * (x-3)
    #Regla de actualizacion
    x = x - alpha * grad
print(x)

#03/02/26
```
