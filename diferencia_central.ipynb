```python
#diferencia hacia adelante
import numpy as np #Importa la libreria numpy para calculos numericos

def f(x): #Define la funcion matematica a analisar
  return x**3 #retorna el valor de la funcion x^3

x = 2 #Punto donde se calculara la derivada
h = 0.001  #Tamaño del paso numerico

derivada = (f(x+h)-f(x))/h #Aplica la formula de diferencia hacia adelante

real = 3*x**2 #Calcula la derivada analitica real

error = abs(real - derivada)  #Calcula el error absoluto

print("Derivada numerica:", derivada ) #Muestra el valor aproximado

print("Derivada real:", real) #Muestra el valor exacto

print("Error: ",error) #Muestra la ddiferencia entre valores
```

```python
#Diferencia central
derivada_2 = (f(x+h)-f(x-h))/(2*h) #Aplica la formula de diferencia central


error = abs(real - derivada_2)  #Calcula el error absoluto


print("Derivada central:", derivada_2) #Muestra el valor central

print("Derivada real:", real) #Resultado exacto

print("Error: ", error)
```

```python
#Aplicacion IA: aproximacion de gradiente

def loss(w): #Define la funcion de perdida
  return w**2  + 3*w + 1 #Define el polinomio

w = 1.5 # Valor del parametro del modelo

h = 0.0001 #Paso pequeño

grad = (loss(w + h) - loss(w - h)) / (2*h) #Aproximacion del gradiente
#gradiente = derivada
real = 2*w + 3 #Gradiente real

print("Gradiente numerico:", grad) #Muestra gradiente aproximado
print("Gradiente real",real) #Muestra el gradiente exacto
```
