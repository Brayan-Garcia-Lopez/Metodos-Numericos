```python
#Aproximacion numerica de una raiz
import math #Permite trabajar con operaciones avanzadas como logaritmos,potencias,raíces cuadradas,trigonometría y más, de forma rápida y precisa.
#Valor real de la raiz de 2
real = math.sqrt(2)
#Valor aproximado
aprox = 1.41
#Calculo del error absoluto
error_absoluto = abs(real - aprox)
#Calculo del error relativo
error_relativo = error_absoluto / real

print(error_absoluto, error_relativo)

#03/02/26
```
