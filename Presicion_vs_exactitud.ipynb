```python
#PRESICION VS EXACTITUD
import numpy as np
#conjunto de mediciones
mediciones = np.array([9.8,9.81,9.79,9.80])
#valor real
valor_real = 9.81

#sacamos la media con mean(media)
media = np.mean(mediciones)
#calculamos el error
error = abs(media - valor_real)

print("Media", media)
print("Error absoluto", error)

#Alta precision (abaja varianza), buena exactitud
```
