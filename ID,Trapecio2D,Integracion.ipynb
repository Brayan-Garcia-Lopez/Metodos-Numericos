```python
#Integral doble basica

import numpy as np

def f(x,y): #Define la funcion de dos variables
     return x + y #Retorna la suma de las variables

a,b = 0, 1 #Define los limeites en el eje x
c,d = 0, 1 #Define los limeites en el eje y

n = 50 #Numero de divisiones en x
m = 50 #Numero de divisiones en y

dx = (b-a)/n #Calcula el tamaño del paso en x
dy = (d-c)/m #Calcula el tamaño del paso en x

integral = 0 #Inicializa el acumulador de la integral

for i in range(n): #Itera sobre las particiones en x
  for j in range(m): ##Itera sobre las particiones en y

      x = a + i*dx #Calcula la cordenada x en la malla
      y = c + j*dy #Calcula la cordenada y en la malla

      integral += f(x,y)*dx*dy #suma el area elemental

print("Integral doble aproximada:",integral) #Imprime el resultado de la integral
```

```python
#Metodo del trapecio en 2d

def f(x,y): #Define la funcion bidimencional
  return x**2 + y**2 #Funcion cuadratica

a,b = 0,1 #Limites en x
c,d = 0,1 #Limites en y

n = 50 #Numero de puntos en x
m = 50 #Numero de puntos en y

x = np.linspace(a,b,n) #Genera puntos en x
y = np.linspace(c,d,m) #Genera puntos en y

dx = (b - a)/(n-1) #Paso en x
dy = (d - c)/(m-1) #Paso en y

integral = 0 #Inicializa el acumulador

for i in range(n): #Itera en x
    for j in range(m): #Itera en y

       peso = 1 #Inicializa el peso

       if i == 0 or i == n-1: #Verifica si esta en borde en x
            peso *= 0.5 #Ajusta peso en borde
       if j == 0 or j == m-1: #Verifica si esta en borde en y
            peso *= 0.5 #Ajusta peso en borde


       integral += peso * f(x[i],y[j]) #Acumula valor ponderado

integral *= dx*dy #Multiplica por el area del elemento

print("Integral (Trapecio 2D):", integral) #Muestra el resultado
```

```python
#Inttegracion tipo

def f(x,y):
  return np.exp(-(x**2 + y**2)) #Calcula e^(-(x^2 + x^2))

a,b = -1,1 #Intervalo en x
c,d = -1,1 #Intervalo en y

n = 100 #Numero de divisiones en x
m = 100 #Numero de divisiones en y

dx = (b-a)/n #Paso x
dy = (d-c)/m #Paso y

integral = 0 #Inicializa el acumulador

for i in range(n): #Itera en x
  for j in range(m): #Itera en y

    x = a + i*dx #Calcula punto en x
    y = c + j*dy #Calcula punto en y

    integral += f(x,y)*dx*dy #Acumula contribucion

print("Probabilidad aproximacion",integral) #Muestra resultado
```
