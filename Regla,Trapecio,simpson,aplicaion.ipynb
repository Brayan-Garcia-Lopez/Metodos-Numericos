```python
 #regla del trapecio
import numpy as np #Importa la libreria numpy para calculos matematicos

def f(x): #Define la funcion a integrar
  return x**2 #Retorna el valor de la funcion x^2

a = 0 #Limite inferior de integracion
b = 2 #Limite superior de integracion
n = 100 #Numero de subintervalos

h = (b-a)/n #Calcula el tamaño del paso

x = np.linspace(a,b,n+1) #Genera los puntos de discretizacion del intervalo

y = f(x) #Evalua la funcion en todos los puntos

integral = (h/2)*(y[0]+2*np.sum(y[1:n])+y[n]) #aplica la forma del trapecio compuesto

print("Integral aproximada (Trapecio):", integral) #Muestra los resultados de la integral
```

```python
#regla de simpson

a = 0 #Limite inferior
b = 2 #Limite superior
n = 100 #Numero de sub intervalos (debe ser par)

h = (b-a)/h #Calcula el tamaño del paso

x = np.linspace(a,b,n+1) #Genera los  nodos del intervalo

y = f(x) #Evalua la fucion

suma = y[0] + y[n] #Inicializa la suma con extremos

for i in range(1,n): #Itera sobre los puntos inferiores
  if i % 2 == 0: #Verifica si el indice es par
    suma += 2*y[i] #Aplica coeficiente 2
  else:
    suma += 4*y[i] #Aplica coeficiente 4

integral = (h/3)*suma #Aplica la forma de Simpson

print("Integral aproximada (Simpson):", integral) #Muestra el resultado
```

```python
#Aplicaion IA: Calculo de area bajo la curva
#f(x) = e^x^2

def f(x):
  return np.exp(-x**2)

a = 0 #Limite inferior
b = 1 #Limite superior
n = 100 #Numero de intervalos

h = (b-a)/n #Tamaño del paso

x = np.linspace(a,b,n+1) #Genera puntos

y = f(x) #Evalua la funcion

integral = (h/2)*(y[0]+2*np.sum(y[1:n])+y[n]) #Aplica la fromula del trapecio

print("Area bajo la  curva", integral) #Resultado

#La integracio numerica permite trabajar con funciones complejas
```
