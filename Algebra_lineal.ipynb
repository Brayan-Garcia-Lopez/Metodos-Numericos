```python
#ALGEBRA LINEAL
import numpy as np

#matriz de 2x2
A = np.array([[3,2],
             [1,2]])

#Vector
b = np.array([5,5])

#Resolucion al sistema
x = np.linalg.solve(A,b) #-- linalg: proporciona un conjunto completo de funciones para operaciones de álgebra lineal

print(x)
```
