```python
#tabla de diferencias divididas
import numpy as np

x = np.array([1,2,3]) #define los valores de x conocidos
y = np.array([1,4,9]) #define los valores de y correspondientes

n = len(x) #Calcula el numero de puntos

tabla = np.zeros((n,n)) #Inicializa la tabla de diferencias divididas


tabla[:,0] = y #Llena la primera columna con valores de y

for j in range(1,n): #Itera sobre las columnas de la tabla
  for i in range(n-j): #Itera sobre las filas disponibles
    tabla[i,j] = (tabla[i+1,j-1] - tabla[i,j-1])/(x[i+j] - x[i]) #calcula la diferencia


print("Tabla de diferencias divididas: \n",tabla) #Muestra la tabla
```

```python
#Evaluacion de polinomio de newtoon

x = np.array([1,2,3]) #define los valores de x conocidos
y = np.array([1,4,9]) #define los valores de y correspondientes

n = len(x) #Calcula el numero de puntos

tabla = np.zeros((n,n)) #Inicializa la tabla de diferencias divididas


tabla[:,0] = y #Llena la primera columna con valores de y

for j in range(1,n): #Itera sobre las columnas de la tabla
  for i in range(n-j): #Itera sobre las filas disponibles
    tabla[i,j] = (tabla[i+1,j-1] - tabla[i,j-1])/(x[i+j] - x[i]) #calcula la diferencia


def newton_eval(x_data, tabla, x_val): #Define funcion para evaluar el polinomio
  n = len(x_data) #numero de puntos
  resultado = tabla[0,0] #Inicializa el primer coeficiente

  producto = 1 #Inicializa el producto acumulado

  for i in range(1,n): #Itera sobre terminos del polinomio
    producto *= (x_val - x_data[i-1]) #Calcula el termino (x-x_i)
    resultado += tabla[0,i]*producto #Suma el termino correspondiente

  return resultado #Retorna valor interpolado

valor = newton_eval(x,tabla,2.5) #Evalua el polinomio en x=2.5

print("Valor Interpolado:",valor) #Muestra resultado
```

```python
#Aplicaion IA:Interpolacion de datos

x = np.array([0,1,2,3]) #define los valores de x conocidos
y = np.array([1,2,0,5]) #define los valores de y correspondientes

n = len(x) #Calcula el numero de puntos

tabla = np.zeros((n,n)) #Inicializa la tabla de diferencias divididas


tabla[:,0] = y #Llena la primera columna con valores de y

for j in range(1,n): #Itera sobre las columnas de la tabla
  for i in range(n-j): #Itera sobre las filas disponibles
    tabla[i,j] = (tabla[i+1,j-1] - tabla[i,j-1])/(x[i+j] - x[i]) #calcula la diferencia


def newton(x_data, tabla, x_val): #Define funcion para evaluar el polinomio
  n = len(x_data) #numero de puntos
  resultado = tabla[0,0] #Inicializa el primer coeficiente
  producto = 1

  for i in range(1,n): #Itera sobre terminos del polinomio
    producto *= (x_val - x_data[i-1]) #Calcula el termino (x-x_i)
    resultado += tabla[0,i]*producto #Suma el termino correspondiente

  return resultado #Retorna valor interpolado

x_interp = 1.5 #punto donde se quiere estimar el valor

y_interp = newton(x,tabla,x_interp) #Interpolacion

print("Valor estimado: ",y_interp) #Resultado final
```

```python
#Implementacion basica de langrage

import numpy as np

x = np.array([0,1,2,3]) #define los valores de x conocidos
y = np.array([1,2,0,5]) #define los valores de y correspondientes


def langrage(x_data, y_data, x_val): #Define funcion de interpolacion de langrage
  n = len(x_data) #Obtiene el numero de puntos
  resultado = 0 #Inicializa el resultado del polinomio

  for i in range(n):#Itera sobre cada punto
    L = 1 #Inicializa el polinomio base l_i

    for j in range(n): #Itera sobre todos los puntos para construir l_i
      if j != i: #Evita dividir entre 0 cuando i=j
        L *= (x_val - x_data[j])/(x_data[i] - x_data[j]) #Calcula el polinomio

    resultado += y_data[i] #Acumula el termino correspondiente al polinomio
  return resultado #Retorna el valor interpolado

valor = langrage(x,y,2.5) #Evalua el polinomio en x=2.5

print("Valor interpolado: ",valor)

```

```python
#Aplicacion IA: Reconstruccion de datos

x = np.array([0,1,2,3]) #Datos en x (Pueden ser tiempos o features)
y = np.array([1,3,2,5]) #Datos observados(Pueden tener ruido)

def langrage(x_data, y_data, x_val): #Define funcion de interpolacion de langrage
  n = len(x_data) #Obtiene el numero de puntos
  resultado = 0 #Inicializa el resultado del polinomio

  for i in range(n):#Itera sobre cada punto
    L = 1 #Inicializa el polinomio base l_i

    for j in range(n): #Itera sobre todos los puntos para construir l_i
      if j != i: #Evita dividir entre 0 cuando i=j
        L *= (x_val - x_data[j])/(x_data[i] - x_data[j]) #Calcula el polinomio

    resultado += L * y_data[i] #Acumula el termino correspondiente al polinomio. Fixed the missing L * here.

  return resultado #Retorna el valor interpolado

x_interp = 1.5


y_interp = langrage(x,y,x_interp)

print("Valor estimado",y_interp)

#Permite construir modelos exactos apartir de datos es intuitiva y directa util en datset pequeños base para metodos mas avanzados
```

```python
#interpolacion segmentada(spline)
import numpy as np
from scipy.interpolate import CubicSpline

x = np.array([0,1,2,3])#deine los puntos en el eje x
y = np.array([1,3,2,5]) #Define los valores correspondientes en y

scpline = CubicSpline(x,y) #Construye scpline cubico que interpola los puntos

x_interp = np.linspace (0,3,100)

y_interp = scpline(x_interp)

print("Valores interpolados:",y_interp)
```

```python
#Regresion lineal
x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])

coef = np.polyfit(x,y,1)

m = coef[0]
b = coef[1]

y_pred = m*x + b

print("Pendiente:",m)
print("Intercepto:",b)
print("Prediccion:",y_pred)
```

```python
#Regresion lineal ccon minimos cuadrados(Manual)
import numpy as np

x = np.array([1,2,3,4,5]) #Define los datos de entrada
y = np.array([2,4,5,4,5]) #define los datos observados

X = np.vstack([np.ones(len(x)), x]).T #Construye la matriz de diseño con columnas

XT = X.T #Calcula .a transpuesta de la mariz X

beta = np.linalg.inv(XT.dot(X)).dot(XT).dot(y) #Aplica la formula de minimos cuadrados

b = beta[0] #Extrae el intercepto
m = beta[1] #Extrae la pendiente

y_pred = m*x+b #Calcula los valores estimados para el modelo

print("Pendiente",m) #Muestra la pendeinte
print("intercepto",b) #Muestra el intercepto
print("Prediccion",y_pred) #Muestra la prediccion
```

```python
#Ajuste polinomial
x = np.array([1,2,3,4,5]) #Define los datos de entrada
y = np.array([2,4,5,4,5]) #define los datos observados

coef = np.polyfit(x,y,2)  #Ajuste polinomial de grado 2

a = coef[0] # Coeficiente cuadratico
b = coef[1] # Coeficinte lineal
c = coef[2] # Termino Independiente

y_pred = a*x**2 + b*x + c #Calcula valores predichos

print("Coeficientes",coef) #Muestra coeficientes
print("Prediccion",y_pred) #Muestra el ajuste
```

```python
#Error cuadratico medio(MSE)

x = np.array([1,2,3,4,5]) #Define los datos de entrada
y = np.array([2,4,5,4,5]) #define los datos observados

coef = np.polyfit(x,y,1) #Ajuste  lineal

y_pred = coef[0]*x + coef[1] #Predicciones

error = np.mean((y - y_pred)**2) #calculla el error cuadratico

print("Error MSE", error) #Muestra el error
```
