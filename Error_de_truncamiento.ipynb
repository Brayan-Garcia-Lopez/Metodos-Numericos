```python
#ERROR DE TRUNCAMIENTO
import math
#valor donde se evaluara la funcion
x = 1
#Aproximacion usando los primeros 3 términos de la serie
aprox = 1 + x + (x**2)/2
# Valor real usando la función exponencial
real = math.exp(x)

#calculo del error
error = abs(real - aprox)

print("Aprox:", aprox)
print("Real:", real)
print("Error:", error)
```
